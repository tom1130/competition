```python
import cv2
import pandas as pd
from PIL import Image
from tqdm import tqdm
```

```python
data = pd.read_csv('/content/drive/MyDrive/commit_folder/competition/tour_ai/data/train.csv')
```

```python
base_dir = '/content/drive/MyDrive/commit_folder/competition/tour_ai/data'
```

```python
img_paths = data['img_path'].values
```

```python
for img_path in tqdm(img_paths):
  img_path = base_dir+img_path[1:]
  image = Image.open(img_path)
  image_resize = image.resize((224,224), Image.HAMMING)
  image_resize.save(img_path.replace('train','compress_train'))
```
